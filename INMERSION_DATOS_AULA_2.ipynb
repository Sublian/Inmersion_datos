```python
#Desafio 2
"""
Se necesita calcular el precio por metro cuadrado, para ello se necesita agrupar por barrios la suma de la columnas identificadas como "PrecioMillon" y "Area"
"""
BarrioAcumulado =inmuebles[['Barrio','Area','Precio_Millon']].groupby(['Barrio']).sum()
BarrioAcumulado['Precio por mt2']= BarrioAcumulado['Precio_Millon']/BarrioAcumulado['Area']
BarrioAcumulado.head()
```

Ahora necesitamos cambiar la apariencia o visualizacion de los datos, para que solo sea visible la columna Barrio junto con Precio por mt2, para ello usamos el siguiente codigo:

```python
BarrioValorMT2=BarrioAcumulado.reset_index()[['Barrio', 'Precio por mt2']]
BarrioValorMT2.sort_values(by=['Precio por mt2'], ascending=False).tail(10)
```

//en el caso anterior usamos tail() solo para mostrar los ultimos datos

Ahora generaremos graficos para mostrar la data de forma mas eficiente al publico, para ello tomamos en cuenta solo el TOP 10 de precios mayores por metro cuadrado y TOP 10 de precios menores por metros cuadrados
para crear el primer grafico TOP precios mayores crearemos el dataframe con el siguiente codigo

```python
#obtenemos los 10primeros
TOP10MayorPrecio=BarrioValorMT2.sort_values(by=['Precio por mt2'], ascending=False).head(10)

#graficamos
plt.figure(figsize=(12,5))
plt.barh(TOP10MayorPrecio['Barrio'], TOP10MayorPrecio['Precio por mt2'], color='green')
#invertimos los axis en y para que muestre de mayor a menor
plt.gca().invert_yaxis()
plt.title('TOP 10 de Barrios por mayor Precio/MT2')
plt.xlabel('Precio por mt2')
plt.ylabel('Barrios')

#para que muestre el valor exacto para cada campo
for valor, categoria in zip(TOP10MayorPrecio['Precio por mt2'], TOP10MayorPrecio['Barrio']):
  plt.text(valor, categoria, str('%.2f' %valor), ha='left', va='center')
```

El codigo anterior genera una grafica de barras horizontales mostrando los mayores precios por metro cuadrado en orden descendente, con color verde y etiquetas para su facil comprension con dos decimales como maximo

Ahora generaremos la siguiente grafica TOP 10 de Barrios por menor Precio/MT2, el cual tiene un codigo similar

```python
#obtenemos los 10primeros
TOP10MenorPrecio=BarrioValorMT2.sort_values(by=['Precio por mt2']).head(10)

#graficamos
plt.figure(figsize=(12,5))
plt.barh(TOP10MenorPrecio['Barrio'], TOP10MenorPrecio['Precio por mt2'], color='blue')
#invertimos los axis en y para que muestre de mayor a menor
plt.gca().invert_yaxis()
plt.title('TOP 10 de Barrios por Menor Precio/MT2')
plt.xlabel('Precio por mt2')
plt.ylabel('Barrios')

#para que muestre el valor exacto para cada campo
for valor, categoria in zip(TOP10MenorPrecio['Precio por mt2'], TOP10MenorPrecio['Barrio']):
  plt.text(valor, categoria, str('%.2f' %valor), ha='left', va='center')
```

Ahora para mejorar la presentacion se colocaran las dos graficas en una misma imagen para facilitar la comprension

```python
fig, axs =plt.subplots(1,2, figsize=(12,5))
#primer grafico TOP 10 mayor
axs[0].barh(TOP10MayorPrecio['Barrio'], TOP10MayorPrecio['Precio por mt2'], color= 'green')
axs[0].invert_yaxis()
axs[0].set_xlabel('Precio Promedio')
axs[0].set_ylabel('Barrio')
axs[0].set_title('TOP 10 Mayor Precio/MTS2', fontweight='bold')
for valor, categoria in zip(TOP10MayorPrecio['Precio por mt2'], TOP10MayorPrecio['Barrio']):
  axs[0].text(valor, categoria, str('%.2f' %valor), ha='left', va='center')

#segundo grafico TOP 10 menor
axs[1].barh(TOP10MenorPrecio['Barrio'], TOP10MenorPrecio['Precio por mt2'], color= 'blue')
axs[1].invert_yaxis()
axs[1].set_xlabel('Precio Promedio')
axs[1].set_ylabel('Barrio')
axs[1].set_title('TOP 10 Menor Precio/MTS2', fontweight='bold')
for valor, categoria in zip(TOP10MenorPrecio['Precio por mt2'], TOP10MenorPrecio['Barrio']):
  axs[1].text(valor, categoria, str('%.2f' %valor), ha='left', va='center')

#ajustar espacio entre graficos
plt.tight_layout()

#mostrar figura
plt.show()
```

**Primer desafio**
Ahora analizaremos las siguientes propiedades **Casa, Apartamento y Finca** por separado a fin de obtener informacion concluyente para cada caso

Comenzamos con la distribucion de frecuencias para las casas, para luego mostrar un grafico
Usaremos el siguiente codigo:

```python
plt.figure(figsize=(10,6))
grafica1 = sns.histplot(data=inmuebles[inmuebles['Tipo']=='Casa'], x='Precio_Millon', kde=True, hue='Tipo')
grafica1.set(title='Distribucion de Precio en Millones de los inmuebles de tipo Casa en Bogota')
plt.xlabel('Precio en Millones de ~Colombianos')
plt.ylabel('Frecuencia')
```

Ahora vamos a crear un gráfico combinado en el cual se mostrará los tres tipos de propiedades seleccionados que son:** Casa, Apartamento y Finca**. Para ello escribo el siguiente código:

```python
arrayTipo=['Apartamento','Casa','Finca']
colores = ['blue', 'green', 'red']
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
for i, tipo in enumerate(arrayTipo):
    datos_tipo = inmuebles[inmuebles['Tipo'] == tipo]
    sns.histplot(data=datos_tipo, x='Precio_Millon', bins=20, kde=True, ax=axs[i],color=colores[i])
    axs[i].set_xlabel('Precio en Millones')
    axs[i].set_ylabel('Frecuencia')
    axs[i].set_title(f'Histograma de Precios de {tipo}', fontweight='bold', color=colores[i])
plt.show()
```

Se puede observar, cada gráfico posee un título con el mismo color de las columnas y en negrita, el color de las columnas y etiquetas a cada eje de coordenadas.

Ahora lo se realizara  una comparación entre los histogramas de Apartamento y Casas y para ello escribimos el siguiente código:

```python
DataCasaApartamento=inmuebles[inmuebles['Tipo'].isin(['Casa','Apartamento'])]
plt.figure(figsize=(10,5))
grafica=sns.histplot(data=DataCasaApartamento,x='Precio_Millon',kde=True,hue='Tipo')
```

Se observa en dicho gráfico un comportamiento similar de las distribuciones antes de los 2000 millones. Sin embargo, necesito realizar un zoom para ver cómo se comportan las distribuciones para valores superiores a los 2000 millones y para ello escribo el siguiente código:

```python
DataCasaApartamento=inmuebles[inmuebles['Tipo'].isin(['Casa','Apartamento'])]
plt.figure(figsize=(10,5))
plt.xlim((2000,12000))
plt.ylim((0,500))
grafica=sns.histplot(data=DataCasaApartamento,x='Precio_Millon',kde=True,hue='Tipo',bins=20)

#Dicho código nos genera la siguiente gráfica:
```

Se observa en este caso que para valores superiores a los 2000 millones de pesos colombianos las distribuciones se comportan de modo similar.

Ahora realizaré una comparación de Apartamentos y Fincas en el cual aprovecho de realizar un zoom y para ello escribo el siguiente código:

```python
DataCasaFinca=inmuebles[inmuebles['Tipo'].isin(['Finca','Apartamento'])]
plt.figure(figsize=(10,8))
grafica=sns.histplot(data=DataCasaFinca,x='Precio_Millon',hue='Tipo',bins=10)
plt.ylim((0,100))
```

Realizo un zoom debido a como bien puede observarse del gráfico combinado de histogramas, la distribución de frecuencias para las fincas es, notoriamente, distinta en comparación a las distribuciones de Casas y Apartamentos. El código me genera el anterior plot
